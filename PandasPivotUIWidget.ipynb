```python
from IPython.html import widgets

import IPython.display as displayTools
from IPython.display import display

from IPython.display import HTML

import matplotlib.pyplot as plt


from pandas import Series, DataFrame, pivot_table
import pandas as pd
import numpy as np
import numpy

pd.options.display.float_format = '{:20,.2f}'.format

from datetime import datetime
def stamp():
    print datetime.now().strftime('%Y/%m/%d %H:%M:%S')
stamp()
```

```python
url = 'https://raw.github.com/pydata/pandas/master/pandas/tests/data/tips.csv'
d = pd.read_csv(url)
```

```python
class PandasPivotUIWidget:
    # Copyright Balazs Kozma, 2014
    # Use it at your own risk:)
    
    # used to store last pivot setup
    lastRow = None
    lastCol = None
    lastValue = None
    #lastAggFunc = None #it seems that the widget 'cannot release' the handle to the function 
    lastDF = None
    
    def __init__(self, df):
        self.df = df;
    
    def display(self):
        self.displayPivotUI()
     
    def displayPivotUI(self):
        self.wPopup = widgets.PopupWidget(description="Pivot params")
        display(self.wPopup)
        self.setupDisplayChooser()
        self.setupTabs()
        self.showPivot()
        self.wPopup.children = [self.wTabsInput, self.wMultiToggleButton]
        #Without the below lines, Tab labels and Display buttons are not updated before first mouse click
        self.wTabsInput.set_title(15,'needed for proper rendering')
        self.configureDisplayChooser()
    
    def showPivot(self):
        displayTools.clear_output(wait=True)
        if self.wColumns.value == self.wRows.value or \
           self.wColumns.value == self.wValues.value or \
           self.wRows.value == self.wValues.value:
                display(HTML('Column, row, value cannot be the same. No pivot report is available'))
                display(HTML('row = '+ self.wRows.value))
                display(HTML('col = '+ self.wColumns.value))
                display(HTML('value = '+ self.wValues.value))
        else:
            piv = pd.tools.pivot.pivot_table(self.df, index=self.wRows.value, 
                                            columns = self.wColumns.value, 
                                            values = self.wValues.value, 
                                            aggfunc = self.wAggFunc.value)
            if self.statePlot['Plot'].value: 
                piv.plot(kind='barh', title= self.wAggFunc.value_name + ' ' + self.wValues.value)
                f = plt.gcf()
                f.set_size_inches(7,5)
            if self.statePlot['Table'].value:
                display(HTML('<b>' + self.wAggFunc.value_name + ' ' + self.wValues.value + '</b>'))
                display(HTML( piv.to_html())) 
        self.saveLastState()

    def saveLastState(self):
        PivotUI.lastRow = self.wRows.value
        PivotUI.lastCol = self.wColumns.value
        PivotUI.lastValue = self.wValues.value
        #PivotUI.lastAggFunc = self.wAggFunc.value
        
    def doFilterValues(self):
        if self.wValuesNumOnly.value:
            self.valuesMask = self.df.columns.isin(self.df._get_numeric_data().columns)
        else:
            self.valuesMask = [True] * len(self.df.columns)
        self.wValues.values = { c:c for c in self.df.columns[self.valuesMask]}

    def doFilterRows(self):
        if self.wRowsNonNumOnly.value:
            self.rowsMask = ~self.df.columns.isin(self.df._get_numeric_data().columns)
        else:
            self.rowsMask = [True] * len(self.df.columns)
        self.wRows.values = { c:c for c in self.df.columns[self.rowsMask]}

    def doFilterCols(self):
        if self.wColsNonNumOnly.value:
            self.colsMask = ~self.df.columns.isin(self.df._get_numeric_data().columns)
        else:
            self.colsMask = [True] * len(self.df.columns)
        self.wColumns.values = { c:c for c in self.df.columns[self.colsMask]}

    def setupTabs(self):
        self.wRows = widgets.SelectWidget(values=list(self.df.columns))
        self.wColumns = widgets.SelectWidget(values=list(self.df.columns))
        self.wValues = widgets.SelectWidget(values=list(self.df.columns))
        self.wAggFunc = widgets.SelectWidget( values={ 'sum' : numpy.sum, 
                                                       'ave' : numpy.average , 
                                                       'sum abs': lambda x: numpy.sum(numpy.abs(x))}, 
                                              value = numpy.sum)

        self.wRowsNonNumOnly = widgets.CheckboxWidget(description="Filter for non numeric values", value=True)
        self.wColsNonNumOnly = widgets.CheckboxWidget(description="Filter for non numeric values", value=True)
        self.wValuesNumOnly = widgets.CheckboxWidget(description="Filter for numeric values", value=True)
        
        nNumericCols = size(self.df._get_numeric_data().columns)
        nNonNumericCols =  size(self.df.columns) - nNumericCols
        if nNonNumericCols <= 1:
            self.wRowsNonNumOnly.value = False
            self.wColsNonNumOnly.value = False
            self.wRowsNonNumOnly.disabled = True
            self.wColsNonNumOnly.disabled = True
            
        if nNumericCols <= 1:
            self.wValuesNumOnly.value = False
            self.wValuesNumOnly.disabled = True
        
        self.doFilterRows()
        self.doFilterCols()
        self.doFilterValues()
        
        #use the last pivot setup if the dataframe is the same as last time
        if PivotUI.lastDF is None or not PivotUI.lastDF is self.df:
            PivotUI.lastDF = self.df
            self.wRows.value = self.wRows.values.values()[0]
            self.wColumns.value = self.wColumns.values.values()[1]
            self.wValues.value = self.wValues.values.values()[0]
        else:
            self.wRows.value = PivotUI.lastRow
            self.wColumns.value = PivotUI.lastCol
            self.wValues.value = PivotUI.lastValue
            #self.wAggFunc.value = PivotUI.lastAggFunc

       
        self.wRows.on_trait_change(self.showPivot, 'value')
        self.wColumns.on_trait_change(self.showPivot, 'value')
        self.wValues.on_trait_change(self.showPivot, 'value')
        self.wAggFunc.on_trait_change(self.showPivot, 'value')
        
        self.wRowsNonNumOnly.on_trait_change(self.doFilterRows, 'value')
        self.wColsNonNumOnly.on_trait_change(self.doFilterCols, 'value')
        self.wValuesNumOnly.on_trait_change(self.doFilterValues, 'value')
        
        self.wContainerRows = widgets.ContainerWidget(children=[self.wRowsNonNumOnly, self.wRows])
        self.wContainerColumns = widgets.ContainerWidget(children=[self.wColsNonNumOnly, self.wColumns])
        self.wContainerValues = widgets.ContainerWidget(children=[self.wValuesNumOnly, self.wValues])

        self.wTabsInput = widgets.TabWidget(children=[self.wContainerRows,
                                                      self.wContainerColumns,
                                                      self.wContainerValues,
                                                      self.wAggFunc])#,self.wPlt])
        
        self.wTabsInput.set_title(0,'Rows')
        self.wTabsInput.set_title(1,'Columns')
        self.wTabsInput.set_title(2,'Values')
        self.wTabsInput.set_title(3,'AggFunc')

    def setupDisplayChooser(self):
        self.wT0 = widgets.HTMLWidget(value='Display:')
        self.wT0.set_css('margin','3px')
        self.wT1 = widgets.ToggleButtonWidget(description = 'Table', value = True)
        self.wT2 = widgets.ToggleButtonWidget(description = 'Figure', value = False)
        self.wMultiToggleButton = widgets.ContainerWidget(children=[self.wT0, self.wT1, self.wT2])
        self.statePlot = {'Table':self.wT1, 'Plot':self.wT2}
        
        self.wT1.on_trait_change(self.showPivot, 'value')
        self.wT2.on_trait_change(self.showPivot, 'value')
        
    def configureDisplayChooser(self):
        self.wMultiToggleButton.remove_class('vbox')
        self.wMultiToggleButton.add_class('hbox')
        self.wMultiToggleButton.set_css('margin','5px 5px')
```

```python
pUI = PandasPivotUIWidget(d)
pUI.display()
```
